##5.

```python
meses = [
    "Julio",
    "Agosto",
    "Septiembre",
    "Octubre",
    "Noviembre",
    "Diciembre"
]
```

```python
bases = []

for mes in meses:

    carpeta = ruta + mes + "/"

    caracteristicas = pd.read_csv(carpeta + "caracteristicas.csv", sep=";", encoding="latin-1", low_memory=False)
    hogar = pd.read_csv(carpeta + "hogar.csv", sep=";", encoding="latin-1", low_memory=False)
    fuerza = pd.read_csv(carpeta + "fuerza.csv", sep=";", encoding="latin-1", low_memory=False)
    ocupados = pd.read_csv(carpeta + "ocupados.csv", sep=";", encoding="latin-1", low_memory=False)

    fuerza = fuerza.drop(columns=["DPTO","PERIODO","MES","PER","CLASE","AREA","REGIS","FEX_C18","HOGAR"], errors="ignore")
    ocupados = ocupados.drop(columns=["DPTO","PERIODO","MES","PER","CLASE","AREA","REGIS","FEX_C18","HOGAR"], errors="ignore")

    base_mes = caracteristicas.merge(
        hogar,
        on=["DIRECTORIO","SECUENCIA_P"],
        how="left"
    )

    base_mes = base_mes.merge(
        fuerza,
        on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
        how="left"
    )

    base_mes = base_mes.merge(
        ocupados,
        on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
        how="left"
    )

    bases.append(base_mes)
```

```python
geih_2025 = pd.concat(bases, ignore_index=True)
```

```python
columnas_limpias = [col for col in geih_2025.columns if not col.endswith('_x')]

geih_2025 = geih_2025[columnas_limpias]
```

```python
ciiu_maestro = pd.read_excel(ruta + "2025-01-14 Correlativa CIIU Rev4.xlsx", sheet_name='CIIU 2022')
```

```python
print(ciiu_maestro.columns)
```

```python
if 'DESCRIPCION_CIIU' in geih_2025.columns:
    geih_2025 = geih_2025.drop(columns=['DESCRIPCION_CIIU'])
if 'DESCRIPCION_CIIU_x' in geih_2025.columns:
    geih_2025 = geih_2025.drop(columns=['DESCRIPCION_CIIU_x'])

geih_2025 = pd.merge(geih_2025,
                       ciiu_maestro[['RAMA4D_R4', 'DESCRIPCION_CIIU']],
                       on='RAMA4D_R4',
                       how='left')
geih_2025
```

```python
divipola_maestro = pd.read_excel(ruta + "2025-01-14 DIVIPOLA.xlsx", sheet_name='Departamentos', header=9)
```

```python
print(divipola_maestro.columns)
```

```python
geih_2025 = pd.merge(geih_2025,
                     divipola_maestro[['Código', 'Nombre']],
                     left_on='DPTO_y',
                     right_on='Código',
                     how='left')

geih_2025 = geih_2025.rename(columns={'Nombre': 'NOMBRE_DEPARTAMENTO'})
geih_2025
```
